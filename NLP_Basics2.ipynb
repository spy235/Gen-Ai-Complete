```python
import pandas as pd
messages=pd.read_csv('spamcollection/spam.csv',
                    sep='\t',names=["label","message"])
```

```python
messages
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

## using PortStemmer

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

### create bag of words

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=100,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
X
```

## using SnowballStemmer

```python
from nltk.stem import SnowballStemmer
snowball_stem = SnowballStemmer("english")
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    [snowball_stem.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

```python
Y=cv.fit_transform(corpus).toarray()
Y
```

## Using Lemitization

```python
from nltk.stem import WordNetLemmatizer
wordnetLem = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    [wordnetLem.lemmatize(word,pos='v') for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

```python
Z=cv.fit_transform(corpus).toarray()
Z
```

## NGRAM

```python
from nltk.corpus import stopwords
import re
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
import pandas as pd
messages=pd.read_csv('spamcollection/spam.csv',
                    sep='\t',names=["label","message"])
```

```python
corpus = []
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

```python
## Create the Bag OF Words model with ngram 
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
# unigram garm
cv=CountVectorizer(max_features=1000,binary=True,ngram_range=(1,1))
```

```python
X=cv.fit_transform(corpus).toarray()
X
```

```python
cv.vocabulary_
```

```python
# binary garm

cv=CountVectorizer(max_features=1000,binary=True,ngram_range=(1,2))
X=cv.fit_transform(corpus).toarray()
cv.vocabulary_
```

```python
# tri garm
cv=CountVectorizer(max_features=1000,binary=True,ngram_range=(3,3))
X=cv.fit_transform(corpus).toarray()
cv.vocabulary_
```
